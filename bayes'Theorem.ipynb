Bayes' Theorem : It helps us update our guess about something based on new evidance . 

Formula:
P(A|B) = [P(B|A) √ó P(A)] / P(B)
Where:

P(A|B) = Posterior probability (what we want to find)
P(B|A) = Likelihood (probability of evidence given hypothesis)
P(A) = Prior probability (initial belief)
P(B) = Evidence (total probability of observing B)

Medical Diagnosis
Scenario: Testing for a disease

D = Person has disease
+ = Test result is positive

Given information:

P(D) = 0.01 (1% of population has the disease) ‚Äî Prior

P(+|D) = 0.99 (Test is 99% accurate if you have disease) ‚ÄîLikelihood

P(+|D') = 0.05 (5% false positive rate) ‚Äî False alarm



Question: If someone tests positive, what's the probability they actually have the disease? Find P(D|+)



Step 1: Calculate P(+) ‚Äî Total probability of testing positive

P(+) = P(+|D) √ó P(D) + P(+|D') √ó P(D')


P(+) = (0.99 √ó 0.01) + (0.05 √ó 0.99)

P(+) = 0.0099 + 0.0495 = 0.0594

Step 2: Apply Bayes' Theorem

P(D|+) = [P(+|D) √ó P(D)] / P(+)

P(D|+) = [0.99 √ó 0.01] / 0.0594

P(D|+) = 0.0099 / 0.0594

P(D|+) ‚âà 0.167 or 16.7%

Surprising result! Even with a positive test, there's only a 16.7%

chance of actually having the disease (because the disease is rare).


```python


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle

# ============================================================================
# MEDICAL DIAGNOSIS - BAYES' THEOREM
# ============================================================================

print("="*70)
print("MEDICAL DIAGNOSIS USING BAYES' THEOREM")
print("="*70)

# Given Information
P_disease = 0.01          # Prior: 1% of population has disease
P_no_disease = 0.99       # 99% don't have disease
P_pos_given_disease = 0.99    # Sensitivity: 99% accurate if diseased
P_pos_given_no_disease = 0.05 # False positive rate: 5%

print("\nüìã GIVEN INFORMATION:")
print(f"P(D) = {P_disease} (Prior - {P_disease*100:.0f}% have disease)")
print(f"P(D') = {P_no_disease} ({P_no_disease*100:.0f}% don't have disease)")
print(f"P(+|D) = {P_pos_given_disease} (Sensitivity - test accuracy)")
print(f"P(+|D') = {P_pos_given_no_disease} (False positive rate)")

# ============================================================================
# STEP 1: Calculate P(+) - Total Probability of Testing Positive
# ============================================================================

print("\n" + "="*70)
print("STEP 1: CALCULATE P(+) - TOTAL PROBABILITY OF TESTING POSITIVE")
print("="*70)

# Using Law of Total Probability
P_positive = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)

print("\nFormula: P(+) = P(+|D) √ó P(D) + P(+|D') √ó P(D')")
print(f"\nCalculation:")
print(f"  P(+|D) √ó P(D) = {P_pos_given_disease} √ó {P_disease} = {P_pos_given_disease * P_disease}")
print(f"  P(+|D') √ó P(D') = {P_pos_given_no_disease} √ó {P_no_disease} = {P_pos_given_no_disease * P_no_disease}")
print(f"\n  P(+) = {P_pos_given_disease * P_disease} + {P_pos_given_no_disease * P_no_disease}")
print(f"  P(+) = {P_positive}")

# ============================================================================
# STEP 2: Apply Bayes' Theorem to Find P(D|+)
# ============================================================================

print("\n" + "="*70)
print("STEP 2: APPLY BAYES' THEOREM - P(D|+)")
print("="*70)

# Bayes' Theorem
P_disease_given_positive = (P_pos_given_disease * P_disease) / P_positive

print("\nBayes' Theorem: P(D|+) = [P(+|D) √ó P(D)] / P(+)")
print(f"\nCalculation:")
print(f"  Numerator: P(+|D) √ó P(D) = {P_pos_given_disease} √ó {P_disease} = {P_pos_given_disease * P_disease}")
print(f"  Denominator: P(+) = {P_positive}")
print(f"\n  P(D|+) = {P_pos_given_disease * P_disease} / {P_positive}")
print(f"  P(D|+) = {P_disease_given_positive:.4f}")

print("\n" + "="*70)
print("üéØ FINAL RESULT")
print("="*70)
print(f"\nProbability of having disease given positive test:")
print(f"  P(D|+) = {P_disease_given_positive:.4f} or {P_disease_given_positive*100:.2f}%")
print(f"\n‚ö†Ô∏è  SURPRISING INSIGHT:")
print(f"  Even with 99% test accuracy and a positive result,")
print(f"  there's only a {P_disease_given_positive*100:.1f}% chance of actually having the disease!")
print(f"\n  Why? Because the disease is RARE (only {P_disease*100:.0f}% prevalence)")

# ============================================================================
# SIMULATION: Test 10,000 People
# ============================================================================

print("\n" + "="*70)
print("SIMULATION: TESTING 10,000 PEOPLE")
print("="*70)

np.random.seed(42)
n_people = 10000

# Generate true disease status
true_disease = np.random.binomial(1, P_disease, n_people)
n_actually_diseased = true_disease.sum()
n_actually_healthy = n_people - n_actually_diseased

print(f"\nPopulation of {n_people:,} people:")
print(f"  Actually have disease: {n_actually_diseased} ({n_actually_diseased/n_people*100:.1f}%)")
print(f"  Actually healthy: {n_actually_healthy} ({n_actually_healthy/n_people*100:.1f}%)")

# Generate test results based on disease status
test_results = np.zeros(n_people)

for i in range(n_people):
    if true_disease[i] == 1:  # Person has disease
        # Test positive with probability P(+|D) = 0.99
        test_results[i] = 1 if np.random.random() < P_pos_given_disease else 0
    else:  # Person is healthy
        # Test positive with probability P(+|D') = 0.05 (false positive)
        test_results[i] = 1 if np.random.random() < P_pos_given_no_disease else 0

# Calculate results
n_positive_tests = test_results.sum()
n_negative_tests = n_people - n_positive_tests

# True positives: Have disease AND test positive
true_positives = ((true_disease == 1) & (test_results == 1)).sum()

# False positives: Healthy BUT test positive
false_positives = ((true_disease == 0) & (test_results == 1)).sum()

# True negatives: Healthy AND test negative
true_negatives = ((true_disease == 0) & (test_results == 0)).sum()

# False negatives: Have disease BUT test negative
false_negatives = ((true_disease == 1) & (test_results == 0)).sum()

print(f"\nTest Results:")
print(f"  Total positive tests: {int(n_positive_tests)}")
print(f"  Total negative tests: {int(n_negative_tests)}")

print(f"\nConfusion Matrix:")
print(f"  True Positives (TP):  {true_positives} - Have disease, test positive ‚úì")
print(f"  False Positives (FP): {false_positives} - Healthy, test positive ‚úó")
print(f"  True Negatives (TN):  {true_negatives} - Healthy, test negative ‚úì")
print(f"  False Negatives (FN): {false_negatives} - Have disease, test negative ‚úó")

# Calculate precision (Positive Predictive Value)
precision = true_positives / n_positive_tests if n_positive_tests > 0 else 0

print(f"\nüìä KEY METRIC:")
print(f"  Precision (PPV) = TP / (TP + FP)")
print(f"  Precision = {true_positives} / {int(n_positive_tests)}")
print(f"  Precision = {precision:.4f} or {precision*100:.2f}%")

print(f"\n‚úÖ Simulation confirms Bayes' Theorem!")
print(f"  Theoretical P(D|+) = {P_disease_given_positive*100:.2f}%")
print(f"  Simulated Precision = {precision*100:.2f}%")

# ============================================================================
# VISUALIZATION 1: Confusion Matrix Heatmap
# ============================================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Confusion Matrix
confusion_matrix = np.array([
    [true_positives, false_negatives],
    [false_positives, true_negatives]
])

ax1 = axes[0, 0]
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Test +', 'Test -'],
            yticklabels=['Disease', 'Healthy'],
            ax=ax1, cbar_kws={'label': 'Count'})
ax1.set_title('Confusion Matrix (10,000 People)', fontsize=14, fontweight='bold')
ax1.set_ylabel('True Status', fontsize=12)
ax1.set_xlabel('Test Result', fontsize=12)

# ============================================================================
# VISUALIZATION 2: Population Breakdown
# ============================================================================

ax2 = axes[0, 1]
categories = ['True\nPositive', 'False\nPositive', 'True\nNegative', 'False\nNegative']
values = [true_positives, false_positives, true_negatives, false_negatives]
colors = ['#2ecc71', '#e74c3c', '#3498db', '#f39c12']

bars = ax2.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
ax2.set_ylabel('Count', fontsize=12)
ax2.set_title('Test Results Breakdown', fontsize=14, fontweight='bold')
ax2.set_ylim(0, max(values) * 1.1)

# Add value labels on bars
for bar, value in zip(bars, values):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(value)}',
            ha='center', va='bottom', fontsize=11, fontweight='bold')

# ============================================================================
# VISUALIZATION 3: Probability Comparison
# ============================================================================

ax3 = axes[1, 0]
prob_labels = ['Prior\nP(D)', 'Test\nAccuracy\nP(+|D)', 'Posterior\nP(D|+)']
prob_values = [P_disease * 100, P_pos_given_disease * 100, P_disease_given_positive * 100]
prob_colors = ['#e67e22', '#9b59b6', '#e74c3c']

bars = ax3.bar(prob_labels, prob_values, color=prob_colors, edgecolor='black', linewidth=1.5)
ax3.set_ylabel('Probability (%)', fontsize=12)
ax3.set_title('Probability Comparison', fontsize=14, fontweight='bold')
ax3.set_ylim(0, 105)
ax3.axhline(y=50, color='gray', linestyle='--', alpha=0.5, label='50%')

# Add value labels
for bar, value in zip(bars, prob_values):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 2,
            f'{value:.1f}%',
            ha='center', va='bottom', fontsize=11, fontweight='bold')

# ============================================================================
# VISUALIZATION 4: Visual Explanation
# ============================================================================

ax4 = axes[1, 1]
ax4.set_xlim(0, 10)
ax4.set_ylim(0, 10)
ax4.axis('off')
ax4.set_title('Visual Explanation of the Paradox', fontsize=14, fontweight='bold')

# Population representation
text_y = 9
ax4.text(5, text_y, 'Out of 10,000 people:', ha='center', fontsize=12, fontweight='bold')

text_y -= 1
ax4.text(5, text_y, f'‚Ä¢ {n_actually_diseased} actually have disease (1%)', ha='center', fontsize=11)
ax4.text(5, text_y - 0.5, f'  ‚Üí {true_positives} test positive (caught!)', ha='center', fontsize=10, color='green')
ax4.text(5, text_y - 1.0, f'  ‚Üí {false_negatives} test negative (missed)', ha='center', fontsize=10, color='orange')

text_y -= 2.5
ax4.text(5, text_y, f'‚Ä¢ {n_actually_healthy} are healthy (99%)', ha='center', fontsize=11)
ax4.text(5, text_y - 0.5, f'  ‚Üí {true_negatives} test negative (correct!)', ha='center', fontsize=10, color='blue')
ax4.text(5, text_y - 1.0, f'  ‚Üí {false_positives} test positive (false alarm!)', ha='center', fontsize=10, color='red')

text_y -= 2.5
ax4.text(5, text_y, f'Total positive tests: {int(n_positive_tests)}', ha='center', fontsize=11, 
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))

text_y -= 1
ax4.text(5, text_y, f'Of these, only {true_positives} truly have disease', ha='center', fontsize=11,
         bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.3))

text_y -= 1.2
ax4.text(5, text_y, f'Precision = {true_positives}/{int(n_positive_tests)} = {precision*100:.1f}%', 
         ha='center', fontsize=12, fontweight='bold', color='red')

plt.tight_layout()
plt.savefig('medical_diagnosis_bayes.png', dpi=300, bbox_inches='tight')
print("\nüìä Visualizations saved as 'medical_diagnosis_bayes.png'")

# ============================================================================
# WHAT IF SCENARIOS
# ============================================================================

print("\n" + "="*70)
print("WHAT IF SCENARIOS: CHANGING DISEASE PREVALENCE")
print("="*70)

prevalences = [0.001, 0.01, 0.05, 0.10, 0.20, 0.50]

print("\nHow does disease prevalence affect P(D|+)?")
print("\nPrevalence | P(D|+) | Interpretation")
print("-" * 50)

for prev in prevalences:
    P_pos = (P_pos_given_disease * prev) + (P_pos_given_no_disease * (1 - prev))
    P_disease_pos = (P_pos_given_disease * prev) / P_pos
    
    print(f"{prev*100:5.1f}%     | {P_disease_pos*100:5.1f}% | ", end="")
    
    if P_disease_pos < 0.30:
        print("Low confidence - many false positives")
    elif P_disease_pos < 0.70:
        print("Moderate confidence")
    else:
        print("High confidence - mostly true positives")

# ============================================================================
# PLOT: Prevalence vs Posterior Probability
# ============================================================================

fig, ax = plt.subplots(figsize=(10, 6))

prevalence_range = np.linspace(0.001, 0.5, 100)
posterior_probs = []

for prev in prevalence_range:
    P_pos = (P_pos_given_disease * prev) + (P_pos_given_no_disease * (1 - prev))
    posterior = (P_pos_given_disease * prev) / P_pos
    posterior_probs.append(posterior * 100)

ax.plot(prevalence_range * 100, posterior_probs, linewidth=3, color='#e74c3c')
ax.axhline(y=50, color='gray', linestyle='--', alpha=0.5, label='50% threshold')
ax.axvline(x=1, color='blue', linestyle='--', alpha=0.5, label='Original (1% prevalence)')

ax.scatter([1], [P_disease_given_positive * 100], color='red', s=200, zorder=5, 
          label=f'Our case: {P_disease_given_positive*100:.1f}%')

ax.set_xlabel('Disease Prevalence (%)', fontsize=12, fontweight='bold')
ax.set_ylabel('P(Disease|Positive Test) %', fontsize=12, fontweight='bold')
ax.set_title('How Disease Prevalence Affects Posterior Probability', 
            fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10)
ax.set_xlim(0, 50)
ax.set_ylim(0, 100)

plt.tight_layout()
plt.savefig('prevalence_vs_posterior.png', dpi=300, bbox_inches='tight')
print("üìä Prevalence analysis saved as 'prevalence_vs_posterior.png'")

# ============================================================================
# KEY INSIGHTS
# ============================================================================

print("\n" + "="*70)
print("üîë KEY INSIGHTS - THE BASE RATE FALLACY")
print("="*70)

print("\n1. HIGH TEST ACCURACY ‚â† HIGH POSITIVE PREDICTIVE VALUE")
print("   - Test is 99% accurate")
print(f"   - But P(D|+) is only {P_disease_given_positive*100:.1f}%")

print("\n2. BASE RATE (PREVALENCE) MATTERS HUGELY")
print(f"   - Disease is rare ({P_disease*100}%)")
print(f"   - So most positive tests are false alarms")
print(f"   - {false_positives} false positives vs {true_positives} true positives")

print("\n3. FALSE POSITIVES DOMINATE WHEN DISEASE IS RARE")
print(f"   - Healthy people: {n_actually_healthy:,}")
print(f"   - 5% test positive: ~{false_positives} false positives")
print(f"   - This swamps the {true_positives} true positives!")

print("\n4. PRACTICAL IMPLICATIONS")
print("   - Don't panic with one positive test")
print("   - Get a second test (retest)")
print("   - Consider risk factors (update prior)")
print("   - Consult medical professional")

print("\n" + "="*70)
print("‚úÖ COMPLETE ANALYSIS FINISHED!")
print("="*70)
plt.show()
```

Email Spam Filter
Problem:
You have 1000 emails:

400 are spam
600 are ham (not spam)

Word "FREE" appears in:

300 spam emails
60 ham emails

A new email contains "FREE". What's the probability it's spam?
Solution:
Given:

P(Spam) = 400/1000 = 0.4
P(Ham) = 600/1000 = 0.6
P("FREE"|Spam) = 300/400 = 0.75
P("FREE"|Ham) = 60/600 = 0.1

Find: P(Spam|"FREE")
Step 1: Calculate P("FREE")
P("FREE") = P("FREE"|Spam) √ó P(Spam) + P("FREE"|Ham) √ó P(Ham)
P("FREE") = 0.75 √ó 0.4 + 0.1 √ó 0.6
P("FREE") = 0.3 + 0.06 = 0.36
Step 2: Apply Bayes' Theorem
P(Spam|"FREE") = [P("FREE"|Spam) √ó P(Spam)] / P("FREE")
P(Spam|"FREE") = [0.75 √ó 0.4] / 0.36
P(Spam|"FREE") = 0.3 / 0.36
P(Spam|"FREE") = 0.833 or 83.3%

```python
import numpy as np
from collections import defaultdict

# Training data
total_emails = 1000
spam_emails = 400
ham_emails = 600

word_in_spam = 300
word_in_ham = 60

# Prior probabilities
P_spam = spam_emails / total_emails
P_ham = ham_emails / total_emails

# Likelihoods
P_free_given_spam = word_in_spam / spam_emails
P_free_given_ham = word_in_ham / ham_emails

# Evidence
P_free = (P_free_given_spam * P_spam) + (P_free_given_ham * P_ham)

# Posterior
P_spam_given_free = (P_free_given_spam * P_spam) / P_free

print("Email Spam Filter Results:")
print(f"P(Spam) = {P_spam:.3f}")
print(f"P('FREE'|Spam) = {P_free_given_spam:.3f}")
print(f"P('FREE'|Ham) = {P_free_given_ham:.3f}")
print(f"\nP(Spam|'FREE') = {P_spam_given_free:.3f} or {P_spam_given_free*100:.1f}%")
print(f"\n‚úâÔ∏è Verdict: Email with 'FREE' is {'SPAM' if P_spam_given_free > 0.5 else 'HAM'}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import pandas as pd

# ============================================================================
# EMAIL SPAM FILTER USING BAYES' THEOREM
# ============================================================================

print("="*70)
print("EMAIL SPAM FILTER USING BAYES' THEOREM")
print("="*70)

# Given Information
total_emails = 1000
spam_emails = 400
ham_emails = 600

word = "FREE"
word_in_spam = 300
word_in_ham = 60

print("\nüìß DATASET INFORMATION:")
print(f"Total emails: {total_emails}")
print(f"  Spam emails: {spam_emails} ({spam_emails/total_emails*100:.0f}%)")
print(f"  Ham emails: {ham_emails} ({ham_emails/total_emails*100:.0f}%)")
print(f"\nWord '{word}' appears in:")
print(f"  Spam: {word_in_spam} out of {spam_emails} emails")
print(f"  Ham: {word_in_ham} out of {ham_emails} emails")

# Calculate Prior Probabilities
P_spam = spam_emails / total_emails
P_ham = ham_emails / total_emails

print("\nüìä PRIOR PROBABILITIES:")
print(f"P(Spam) = {spam_emails}/{total_emails} = {P_spam}")
print(f"P(Ham) = {ham_emails}/{total_emails} = {P_ham}")

# Calculate Likelihoods
P_free_given_spam = word_in_spam / spam_emails
P_free_given_ham = word_in_ham / ham_emails

print("\nüìà LIKELIHOODS:")
print(f"P('{word}'|Spam) = {word_in_spam}/{spam_emails} = {P_free_given_spam}")
print(f"P('{word}'|Ham) = {word_in_ham}/{ham_emails} = {P_free_given_ham}")

# ============================================================================
# STEP 1: Calculate P("FREE") - Total Probability
# ============================================================================

print("\n" + "="*70)
print("STEP 1: CALCULATE P('FREE') - TOTAL PROBABILITY")
print("="*70)

# Using Law of Total Probability
P_free = (P_free_given_spam * P_spam) + (P_free_given_ham * P_ham)

print("\nFormula: P('FREE') = P('FREE'|Spam) √ó P(Spam) + P('FREE'|Ham) √ó P(Ham)")
print(f"\nCalculation:")
print(f"  P('FREE'|Spam) √ó P(Spam) = {P_free_given_spam} √ó {P_spam} = {P_free_given_spam * P_spam}")
print(f"  P('FREE'|Ham) √ó P(Ham) = {P_free_given_ham} √ó {P_ham} = {P_free_given_ham * P_ham}")
print(f"\n  P('FREE') = {P_free_given_spam * P_spam} + {P_free_given_ham * P_ham}")
print(f"  P('FREE') = {P_free}")

# ============================================================================
# STEP 2: Apply Bayes' Theorem
# ============================================================================

print("\n" + "="*70)
print("STEP 2: APPLY BAYES' THEOREM - P(Spam|'FREE')")
print("="*70)

# Bayes' Theorem
P_spam_given_free = (P_free_given_spam * P_spam) / P_free

print("\nBayes' Theorem: P(Spam|'FREE') = [P('FREE'|Spam) √ó P(Spam)] / P('FREE')")
print(f"\nCalculation:")
print(f"  Numerator: P('FREE'|Spam) √ó P(Spam) = {P_free_given_spam} √ó {P_spam} = {P_free_given_spam * P_spam}")
print(f"  Denominator: P('FREE') = {P_free}")
print(f"\n  P(Spam|'FREE') = {P_free_given_spam * P_spam} / {P_free}")
print(f"  P(Spam|'FREE') = {P_spam_given_free:.4f}")

print("\n" + "="*70)
print("üéØ FINAL RESULT")
print("="*70)
print(f"\nProbability email is SPAM given it contains '{word}':")
print(f"  P(Spam|'{word}') = {P_spam_given_free:.4f} or {P_spam_given_free*100:.2f}%")
print(f"\n‚úâÔ∏è VERDICT: Email with '{word}' is highly likely to be SPAM!")

# Also calculate P(Ham|"FREE") for completeness
P_ham_given_free = 1 - P_spam_given_free
print(f"\nProbability email is HAM given it contains '{word}':")
print(f"  P(Ham|'{word}') = {P_ham_given_free:.4f} or {P_ham_given_free*100:.2f}%")

# ============================================================================
# SIMULATION: Generate Email Dataset
# ============================================================================

print("\n" + "="*70)
print("SIMULATION: GENERATING EMAIL DATASET")
print("="*70)

np.random.seed(42)

# Generate email labels
email_labels = np.array(['spam'] * spam_emails + ['ham'] * ham_emails)
np.random.shuffle(email_labels)

# Generate "FREE" word presence based on probabilities
contains_free = np.zeros(total_emails, dtype=bool)

for i in range(total_emails):
    if email_labels[i] == 'spam':
        # Word appears in 75% of spam
        contains_free[i] = np.random.random() < P_free_given_spam
    else:
        # Word appears in 10% of ham
        contains_free[i] = np.random.random() < P_free_given_ham

# Count statistics
n_free_emails = contains_free.sum()
n_free_spam = ((email_labels == 'spam') & contains_free).sum()
n_free_ham = ((email_labels == 'ham') & contains_free).sum()

print(f"\nSimulated Dataset:")
print(f"Total emails with '{word}': {n_free_emails}")
print(f"  Spam emails with '{word}': {n_free_spam}")
print(f"  Ham emails with '{word}': {n_free_ham}")

# Calculate simulated probability
simulated_p_spam_given_free = n_free_spam / n_free_emails if n_free_emails > 0 else 0

print(f"\nüìä COMPARISON:")
print(f"  Theoretical P(Spam|'{word}') = {P_spam_given_free*100:.2f}%")
print(f"  Simulated P(Spam|'{word}') = {simulated_p_spam_given_free*100:.2f}%")
print(f"\n‚úÖ Simulation confirms Bayes' Theorem!")

# ============================================================================
# CONFUSION MATRIX
# ============================================================================

print("\n" + "="*70)
print("CONFUSION MATRIX - EMAIL CLASSIFICATION")
print("="*70)

# For emails containing "FREE"
free_emails = contains_free
true_spam = email_labels == 'spam'

# True Positives: Spam emails with "FREE" (correctly flagged)
TP = (true_spam & free_emails).sum()

# False Positives: Ham emails with "FREE" (incorrectly flagged)
FP = (~true_spam & free_emails).sum()

# True Negatives: Ham emails without "FREE" (correctly not flagged)
TN = (~true_spam & ~free_emails).sum()

# False Negatives: Spam emails without "FREE" (missed)
FN = (true_spam & ~free_emails).sum()

print("\nIf we flag emails with 'FREE' as spam:")
print(f"  True Positives (TP):  {TP} - Spam correctly flagged")
print(f"  False Positives (FP): {FP} - Ham incorrectly flagged")
print(f"  True Negatives (TN):  {TN} - Ham correctly not flagged")
print(f"  False Negatives (FN): {FN} - Spam missed")

# Calculate metrics
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
accuracy = (TP + TN) / total_emails

print(f"\nüìà PERFORMANCE METRICS:")
print(f"  Precision: {precision:.4f} ({precision*100:.2f}%)")
print(f"  Recall: {recall:.4f} ({recall*100:.2f}%)")
print(f"  F1-Score: {f1_score:.4f}")
print(f"  Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")

# ============================================================================
# VISUALIZATION 1: Confusion Matrix Heatmap
# ============================================================================

fig = plt.figure(figsize=(16, 12))

# Confusion Matrix
ax1 = plt.subplot(2, 3, 1)
confusion_matrix = np.array([[TP, FN], [FP, TN]])
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='RdYlGn', 
            xticklabels=['Flagged as Spam', 'Not Flagged'],
            yticklabels=['Actually Spam', 'Actually Ham'],
            ax=ax1, cbar_kws={'label': 'Count'})
ax1.set_title('Confusion Matrix\n(Emails with "FREE")', fontsize=14, fontweight='bold')

# ============================================================================
# VISUALIZATION 2: Email Distribution
# ============================================================================

ax2 = plt.subplot(2, 3, 2)
categories = ['Spam', 'Ham']
email_counts = [spam_emails, ham_emails]
colors = ['#e74c3c', '#2ecc71']

bars = ax2.bar(categories, email_counts, color=colors, edgecolor='black', linewidth=2)
ax2.set_ylabel('Number of Emails', fontsize=12)
ax2.set_title('Email Distribution', fontsize=14, fontweight='bold')
ax2.set_ylim(0, max(email_counts) * 1.15)

for bar, count in zip(bars, email_counts):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{count}\n({count/total_emails*100:.0f}%)',
            ha='center', va='bottom', fontsize=11, fontweight='bold')

# ============================================================================
# VISUALIZATION 3: Word "FREE" Distribution
# ============================================================================

ax3 = plt.subplot(2, 3, 3)
word_categories = [f'Spam\nwith "{word}"', f'Ham\nwith "{word}"']
word_counts = [n_free_spam, n_free_ham]
word_colors = ['#c0392b', '#27ae60']

bars = ax3.bar(word_categories, word_counts, color=word_colors, edgecolor='black', linewidth=2)
ax3.set_ylabel('Number of Emails', fontsize=12)
ax3.set_title(f'Emails Containing "{word}"', fontsize=14, fontweight='bold')
ax3.set_ylim(0, max(word_counts) * 1.2)

for bar, count in zip(bars, word_counts):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
            f'{count}',
            ha='center', va='bottom', fontsize=12, fontweight='bold')

# ============================================================================
# VISUALIZATION 4: Probability Comparison
# ============================================================================

ax4 = plt.subplot(2, 3, 4)
prob_labels = ['Prior\nP(Spam)', 'Likelihood\nP("FREE"|Spam)', 'Posterior\nP(Spam|"FREE")']
prob_values = [P_spam * 100, P_free_given_spam * 100, P_spam_given_free * 100]
prob_colors = ['#3498db', '#9b59b6', '#e74c3c']

bars = ax4.bar(prob_labels, prob_values, color=prob_colors, edgecolor='black', linewidth=2)
ax4.set_ylabel('Probability (%)', fontsize=12)
ax4.set_title('Probability Evolution via Bayes', fontsize=14, fontweight='bold')
ax4.set_ylim(0, 100)
ax4.axhline(y=50, color='gray', linestyle='--', alpha=0.5, label='50% threshold')

for bar, value in zip(bars, prob_values):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 2,
            f'{value:.1f}%',
            ha='center', va='bottom', fontsize=11, fontweight='bold')

# ============================================================================
# VISUALIZATION 5: Bayes Calculation Breakdown
# ============================================================================

ax5 = plt.subplot(2, 3, 5)
ax5.axis('off')
ax5.set_title('Bayes Theorem Breakdown', fontsize=14, fontweight='bold')

text_y = 0.9
line_height = 0.08

# Title
ax5.text(0.5, text_y, 'Step-by-Step Calculation', ha='center', fontsize=13, 
         fontweight='bold', transform=ax5.transAxes)
text_y -= line_height * 1.5

# Prior
ax5.text(0.5, text_y, f'Prior: P(Spam) = {P_spam}', ha='center', fontsize=11,
         transform=ax5.transAxes,
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))
text_y -= line_height * 1.2

# Likelihood
ax5.text(0.5, text_y, f'Likelihood: P("FREE"|Spam) = {P_free_given_spam}', ha='center', fontsize=11,
         transform=ax5.transAxes,
         bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))
text_y -= line_height * 1.5

# Evidence calculation
ax5.text(0.5, text_y, 'Evidence Calculation:', ha='center', fontsize=11,
         fontweight='bold', transform=ax5.transAxes)
text_y -= line_height

ax5.text(0.5, text_y, f'P("FREE") = {P_free_given_spam} √ó {P_spam} + {P_free_given_ham} √ó {P_ham}',
         ha='center', fontsize=10, transform=ax5.transAxes)
text_y -= line_height

ax5.text(0.5, text_y, f'P("FREE") = {P_free}', ha='center', fontsize=10,
         fontweight='bold', transform=ax5.transAxes,
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))
text_y -= line_height * 1.5

# Posterior calculation
ax5.text(0.5, text_y, 'Posterior Calculation:', ha='center', fontsize=11,
         fontweight='bold', transform=ax5.transAxes)
text_y -= line_height

ax5.text(0.5, text_y, f'P(Spam|"FREE") = ({P_free_given_spam} √ó {P_spam}) / {P_free}',
         ha='center', fontsize=10, transform=ax5.transAxes)
text_y -= line_height

ax5.text(0.5, text_y, f'P(Spam|"FREE") = {P_spam_given_free:.3f} or {P_spam_given_free*100:.1f}%',
         ha='center', fontsize=11, fontweight='bold', color='red',
         transform=ax5.transAxes,
         bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.5))

# ============================================================================
# VISUALIZATION 6: Performance Metrics
# ============================================================================

ax6 = plt.subplot(2, 3, 6)
metrics = ['Precision', 'Recall', 'F1-Score', 'Accuracy']
values = [precision * 100, recall * 100, f1_score * 100, accuracy * 100]
metric_colors = ['#e67e22', '#3498db', '#9b59b6', '#2ecc71']

bars = ax6.barh(metrics, values, color=metric_colors, edgecolor='black', linewidth=2)
ax6.set_xlabel('Score (%)', fontsize=12)
ax6.set_title('Classifier Performance Metrics', fontsize=14, fontweight='bold')
ax6.set_xlim(0, 105)
ax6.axvline(x=50, color='gray', linestyle='--', alpha=0.5)

for bar, value in zip(bars, values):
    width = bar.get_width()
    ax6.text(width + 2, bar.get_y() + bar.get_height()/2.,
            f'{value:.1f}%',
            ha='left', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.savefig('spam_filter_bayes_analysis.png', dpi=300, bbox_inches='tight')
print("\nüìä Visualization saved as 'spam_filter_bayes_analysis.png'")

# ============================================================================
# MULTI-WORD NAIVE BAYES CLASSIFIER
# ============================================================================

print("\n" + "="*70)
print("EXTENDED: MULTI-WORD NAIVE BAYES CLASSIFIER")
print("="*70)

# Define more words with their frequencies
words_data = {
    'FREE': {'spam': 300, 'ham': 60},
    'MONEY': {'spam': 250, 'ham': 30},
    'WIN': {'spam': 200, 'ham': 20},
    'CLICK': {'spam': 280, 'ham': 50},
    'MEETING': {'spam': 30, 'ham': 250},
    'REPORT': {'spam': 20, 'ham': 200},
    'URGENT': {'spam': 180, 'ham': 80}
}

# Calculate likelihoods for all words
word_likelihoods = {}
for word, counts in words_data.items():
    word_likelihoods[word] = {
        'spam': counts['spam'] / spam_emails,
        'ham': counts['ham'] / ham_emails
    }

print("\nüìù WORD LIKELIHOODS:")
print(f"{'Word':<12} P(word|Spam)  P(word|Ham)")
print("-" * 45)
for word in word_likelihoods:
    p_spam = word_likelihoods[word]['spam']
    p_ham = word_likelihoods[word]['ham']
    print(f"{word:<12} {p_spam:>7.3f}       {p_ham:>7.3f}")

def classify_email(words, verbose=True):
    """Classify email using Naive Bayes with multiple words"""
    # Start with priors
    score_spam = P_spam
    score_ham = P_ham
    
    if verbose:
        print(f"\n  Starting priors:")
        print(f"    P(Spam) = {P_spam}")
        print(f"    P(Ham) = {P_ham}")
    
    # Multiply by likelihood of each word
    for word in words:
        if word in word_likelihoods:
            score_spam *= word_likelihoods[word]['spam']
            score_ham *= word_likelihoods[word]['ham']
            
            if verbose:
                print(f"\n  After seeing '{word}':")
                print(f"    Spam score: √ó {word_likelihoods[word]['spam']:.3f} = {score_spam:.6f}")
                print(f"    Ham score:  √ó {word_likelihoods[word]['ham']:.3f} = {score_ham:.6f}")
    
    # Normalize
    total = score_spam + score_ham
    prob_spam = score_spam / total
    prob_ham = score_ham / total
    
    if verbose:
        print(f"\n  Final (normalized):")
        print(f"    P(Spam|words) = {prob_spam:.4f} ({prob_spam*100:.2f}%)")
        print(f"    P(Ham|words) = {prob_ham:.4f} ({prob_ham*100:.2f}%)")
    
    return prob_spam, prob_ham

# Test with different email examples
print("\n" + "="*70)
print("TESTING CLASSIFIER WITH DIFFERENT EMAILS")
print("="*70)

test_emails = [
    ['FREE'],
    ['MEETING'],
    ['FREE', 'MONEY', 'WIN'],
    ['MEETING', 'REPORT'],
    ['FREE', 'URGENT'],
    ['CLICK', 'FREE', 'MONEY', 'WIN']
]

print("\nüì® Email Classification Results:\n")

results = []
for i, email_words in enumerate(test_emails, 1):
    print(f"\n{'='*60}")
    print(f"Email {i}: {email_words}")
    print('='*60)
    
    prob_spam, prob_ham = classify_email(email_words, verbose=True)
    
    prediction = 'SPAM' if prob_spam > prob_ham else 'HAM'
    confidence = max(prob_spam, prob_ham)
    
    print(f"\n  üéØ PREDICTION: {prediction} (Confidence: {confidence*100:.1f}%)")
    
    results.append({
        'email': ', '.join(email_words),
        'spam_prob': prob_spam,
        'ham_prob': prob_ham,
        'prediction': prediction
    })

# ============================================================================
# VISUALIZE MULTI-WORD CLASSIFICATION RESULTS
# ============================================================================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Spam vs Ham probabilities for each test email
email_labels = [f"Email {i+1}" for i in range(len(results))]
spam_probs = [r['spam_prob'] * 100 for r in results]
ham_probs = [r['ham_prob'] * 100 for r in results]

x = np.arange(len(email_labels))
width = 0.35

bars1 = ax1.bar(x - width/2, spam_probs, width, label='P(Spam|words)', 
                color='#e74c3c', edgecolor='black', linewidth=1.5)
bars2 = ax1.bar(x + width/2, ham_probs, width, label='P(Ham|words)',
                color='#2ecc71', edgecolor='black', linewidth=1.5)

ax1.set_ylabel('Probability (%)', fontsize=12, fontweight='bold')
ax1.set_title('Email Classification Results', fontsize=14, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(email_labels, rotation=45, ha='right')
ax1.legend(fontsize=10)
ax1.axhline(y=50, color='gray', linestyle='--', alpha=0.5)
ax1.set_ylim(0, 105)
ax1.grid(axis='y', alpha=0.3)

# Plot 2: Word importance (spam indicator strength)
words = list(word_likelihoods.keys())
spam_likelihood = [word_likelihoods[w]['spam'] for w in words]
ham_likelihood = [word_likelihoods[w]['ham'] for w in words]

# Calculate spam indicator score (higher = more spam-like)
spam_scores = [spam_likelihood[i] / (spam_likelihood[i] + ham_likelihood[i]) * 100 
               for i in range(len(words))]

# Sort by spam score
sorted_indices = np.argsort(spam_scores)[::-1]
sorted_words = [words[i] for i in sorted_indices]
sorted_scores = [spam_scores[i] for i in sorted_indices]

colors_gradient = ['#c0392b' if s > 70 else '#e67e22' if s > 50 else '#f39c12' 
                   for s in sorted_scores]

bars = ax2.barh(sorted_words, sorted_scores, color=colors_gradient, 
                edgecolor='black', linewidth=1.5)
ax2.set_xlabel('Spam Indicator Score (%)', fontsize=12, fontweight='bold')
ax2.set_title('Word Importance for Spam Detection', fontsize=14, fontweight='bold')
ax2.axvline(x=50, color='gray', linestyle='--', alpha=0.5, label='Neutral (50%)')
ax2.set_xlim(0, 105)
ax2.legend(fontsize=10)

for bar, score in zip(bars, sorted_scores):
    width = bar.get_width()
    ax2.text(width + 2, bar.get_y() + bar.get_height()/2.,
            f'{score:.1f}%',
            ha='left', va='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('multi_word_classification.png', dpi=300, bbox_inches='tight')
print("\nüìä Multi-word classification saved as 'multi_word_classification.png'")

# ============================================================================
# KEY INSIGHTS
# ============================================================================

print("\n" + "="*70)
print("üîë KEY INSIGHTS")
print("="*70)

print("\n1. SINGLE WORD ANALYSIS:")
print(f"   - Email with '{word}' has {P_spam_given_free*100:.1f}% chance of being spam")
print(f"   - Prior belief: {P_spam*100:.0f}% spam rate")
print(f"   - Updated belief: {P_spam_given_free*100:.1f}% (doubled!)")

print("\n2. WHY WORD 'FREE' IS A STRONG SPAM INDICATOR:")
print(f"   - Appears in {P_free_given_spam*100:.0f}% of spam emails")
print(f"   - Appears in only {P_free_given_ham*100:.0f}% of ham emails")
print(f"   - Likelihood ratio: {P_free_given_spam/P_free_given_ham:.1f}:1")

print("\n3. COMBINING MULTIPLE WORDS:")
print("   - Naive Bayes multiplies individual word probabilities")
print("   - More spam words ‚Üí higher spam probability")
print("   - Example: ['FREE', 'MONEY', 'WIN'] ‚Üí 99%+ spam probability")

print("\n4. PRACTICAL IMPLICATIONS:")
print("   - Simple but effective spam detection")
print("   - Fast computation (just multiplication)")
print("   - Scales well with more features")
print("   - Used in real email filters (Gmail, Outlook)")

print("\n" + "="*70)
print("‚úÖ COMPLETE SPAM FILTER ANALYSIS FINISHED!")
print("="*70)

plt.show()
```
